```python
# string functions

s = "Meaw Cat Purrr"
# split
print(f"split {s.split()}")
# length
print(f"length {len(s)}")
# count
print(f"count {s.count('r')}")
#  index
print(f"the first occcurance of the letter r is ath the index: {s.index('e')}")
#  upper
print(f"upper {s.upper()}")
#  lower
print(f"lower {s.lower()}")
# starts with
print(f"starts with {s.startswith('M')}")
# ends with
print(f"ends with {s.endswith('r')}")
# capitalize
print(f"capitalize {s.capitalize()}")

```
