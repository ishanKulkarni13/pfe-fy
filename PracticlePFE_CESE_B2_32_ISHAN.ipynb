```python
#Expt 18. Program to find the sum of digits of a number
#Name: Ishan Kulkarni

num = int(input("Enter a number: "))

sum_digits = 0

while num > 0:
    digit = num % 10
    sum_digits += digit
    num = num // 10

print("Sum of digits =", sum_digits)
```

```python
#Expt 25. Program to Put Even and Odd elements in a List into Two Different Lists
#Name: Ishan Kulkarni

numbers = [10, 15, 22, 33, 40, 55]

even_list = []
odd_list = []

for num in numbers:
    if num % 2 == 0:
        even_list.append(num)
    else:
        odd_list.append(num)

print("Even elements:", even_list)
print("Odd elements:", odd_list)
```
