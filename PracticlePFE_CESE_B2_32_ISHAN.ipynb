```python
# 30. Function to Find the Factorial of a Number Using Recursion

def factorial(n):
    if n < 1:
        return 1
    else:
        return n * factorial(n - 1)

num = int(input("Enter a number to find its factorial: "))
result = factorial(num)
print("Factorial of", num, "is", result)
```

```python
# 31. Function to Find the Sum of the Digits of the Number Recursively
def sum_of_digits(n):
    if n == 0:
        return 0
    else:
        return (n % 10) + sum_of_digits(n // 10)

num = int(input("Enter a number to find the sum of its digits: "))
result = sum_of_digits(num)
print("Sum of digits of", num, "is", result)
```

```python
# 32. Create a function named count vowels that accepts a string and returns the number of vowels in the string.

def count_vowels(s):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

string = input("Enter a string to count the number of vowels: ")
vowel_count = count_vowels(string)
print("Number of vowels in the string is:", vowel_count)
```

```python
#Expt 35.	Program to Find all Numbers in a Range which are Perfect Squares and Sum of all Digits in the Number is Less than 10
# Name: Ishan Kulkarni
# Date: 2 april 2026

import math

start = int(input("Enter start: "))
end = int(input("Enter end: "))

result = []
for num in range(start, end + 1):
    if num < 0:
        continue
    root = math.isqrt(num)
    if root * root == num:
        s = 0
        t = num
        while t > 0:
            s = s + (t % 10)
            t = t // 10
        if s < 10:
            result.append(num)

print("Numbers are:", result)
```

```python
#Expt 43.	Program to sort (ascending and descending) a dictionary by value
# Name: Ishan Kulkarni
# Date: 2 april 2026

n = int(input("How many items? "))
d = {}
for i in range(n):
    k = input("Enter key: ")
    v = int(input("Enter value (number): "))
    d[k] = v

asc = dict(sorted(d.items(), key=lambda x: x[1]))
desc = dict(sorted(d.items(), key=lambda x: x[1], reverse=True))

print("Ascending:", asc)
print("Descending:", desc)
```

```python
#Expt 50.	Program to Return a set of elements present in Set A or B, but not both
# Name: Ishan Kulkarni
# Date: 2 april 2026

n1 = int(input("How many elements in set A? "))
A = set()
for i in range(n1):
    A.add(input("Enter element: "))

n2 = int(input("How many elements in set B? "))
B = set()
for i in range(n2):
    B.add(input("Enter element: "))

ans = A.symmetric_difference(B)
print("Result set:", ans)
```

```python
#Expt 51.	Program to Check if two sets have any elements in common.
# Name: Ishan Kulkarni
# Date: 2 april 2026

n1 = int(input("How many elements in set1? "))
set1 = set()
for i in range(n1):
    set1.add(input("Enter element: "))

n2 = int(input("How many elements in set2? "))
set2 = set()
for i in range(n2):
    set2.add(input("Enter element: "))

common = set1.intersection(set2)
if len(common) == 0:
    print("No common elements")
else:
    print("Common elements:", common)
```
